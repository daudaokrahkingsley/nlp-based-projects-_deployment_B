```python
import spacy

# Load the SpaCy model
nlp = spacy.load("en_core_web_sm")

# Sample text
text = "Kingsley is a good Boy."

# Process the text
doc = nlp(text)

# Print tokens and their POS tags
for token in doc:
    print(f"Token: {token.text}, POS: {token.pos_}, Detailed POS: {token.tag_}")
```
